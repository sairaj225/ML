```python
import matplotlib.pyplot as plt

# Create a figure and axes object
fig, ax = plt.subplots()

# Create a graph with two vertices
x = [0, 1]
y = [0, 1]
ax.plot(x, y)

# Add an edge to the graph
ax.plot([0, 0.5], [0, 0.5], color='red')

# Show the plot
plt.show()
```
