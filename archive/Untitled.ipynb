# Extract feature and target variables

```python
X = data.iloc[:, 0:1].values
y = data.iloc[:, 1].values
```

# Checking shape of the features

```python
print(X.shape)
print(y.shape)
```

# Perform train-test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
```

# Creating a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
```

# Training a Linear Regression Model

```python
regressor.fit(X_train, y_train)
```

# Making predictions with the Linear Regression model

```python
y_pred = regressor.predict(X_test)
```

# Linear regression model coefficients
## Equation of line is given by: y = mx + b
## y = 0.81*x + 5.50
## m --> slope
## b --> intercept

```python
# X coefficient
print(regressor.coef_[0])

# Y intercept
print(regressor.intercept_)
```

# Creating a Dataframe for Error Analysis

```python
df_error = pd.DataFrame()
df_error['y_test'] = y_test
df_error['y_pred'] = y_pred
df_error.shape
```

```python
df_error
```

# Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

# Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

# Root Mean Squared Error

```python
import numpy as np
mse = mean_squared_error(y_test, y_pred)

np.sqrt(mse)
```

# Calculating the R-squared (R2) score

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
